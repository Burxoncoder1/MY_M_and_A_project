```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

```python
df1 = pd.read_csv("https://storage.googleapis.com/qwasar-public/only_wood_customer_us_1.csv")
df1['Gender'].unique()  #Saralangan qiymatlarni chop etish
genders = {'0':'Male', '1':'Female', 'F':'Female', 'M':'Male'} #Unique qilingan qiymatlarga gender yuklash
df1['Gender'] = df1['Gender'].replace(genders)   #Yangilangan genderslarni eski Gender bilan replace() orqali almashtirish
df1['FirstName'] = df1['FirstName'].str.replace('\W','')  #Regular expression editor orqali  \W	Har qanday so'zsiz belgi
df1['FirstName'] = df1['FirstName'].str.title()   #Firstname ni title orqali bosh xarfini kattalashtirib tartibga keltiramiz
df1['LastName'] = df1['LastName'].str.replace('\W','')
df1['LastName'] = df1['LastName'].str.title()  ##LastNameni ham tartibga keltirib olamiz
df1['UserName'] = df1['UserName'].str.replace('\W','')
df1['UserName'] = df1['UserName'].str.title()   ##UserNameni ham quydagicha tartiblab olamiz
df1['Email'] = df1['Email'].str.lower()
df1['Age'].unique()
df1['City'].unique()  ##City qiymatlarini tekshirib olamiz va kamchiliklarini topamiz
df1['City'] = df1['City'].str.replace('_',' ')
df1['City'] = df1['City'].str.replace('-',' ')    #Kamchiliklari New_york pastki chiziqni -- chiziqqa replace orqali almashtirdik
df1['City'].unique()   ##City qiymatlarini tekshirib olamiz va kamchiliklarini topamiz
df1['City'] = df1['City'].str.title()    ##Kamchiliklardan biri shaxar nomlari istalgan holatda har xil ularni title( func orqali) faqat bosh xarflarini katta qilib tozalab olamiz
df1['Country'] = 'USA'   ## Country da esa Nan va raqamlar kop edi ularni ham USA ga ozgartirib oldik
df1.head(10)
```

2 --DATAFRAME

```python
df2 = pd.read_csv("https://storage.googleapis.com/qwasar-public/only_wood_customer_us_2.csv",sep = ';',header=None, names=['age','city','gender','name','email'])
# df2['age'][60:100]   ##Age ni toza va tozamasligini tekshirib olamiz
df2['age'] = df2['age'].str.replace('\D','')  #(Year) (yo) ga oxshash xatolarni regular expressns orqali togirlab olamiz
df2['age'].unique()
df2['city'] = df2['city'].str.replace('_',' ')
df2['city'] = df2['city'].str.replace('-',' ')
df2['city'] = df2['city'].str.title()
df2['city'].unique()
df2['gender'].unique()  #Saralangan qiymatlarni chop etish
genderss = {'0':'Male', '1':'Female', 'F':'Female', 'M':'Male'} #Unique qilingan qiymatlarga gender yuklash
df2['gender'] = df2['gender'].replace(genderss)   #Yangilangan genderslarni eski Gender bilan replace() orqali almashtirish
df2['gender'].unique()
df2['name'] = df2['name'].str.replace('\W',' ')
df2['name'] = df2['name'].str.title()
name_df = df2['name'].str.split(expand = True)
name_df[2] == None
name_df2 = df2['name'].str.split(expand=True)
df2['first_name'],df2['last_name'] = name_df2[0],name_df2[1]
df2.drop('name', axis = 1, inplace = True)
df2['email'] = df2['email'].str.lower()
df2.head(10)
```

```python
df3 = pd.read_csv("https://storage.googleapis.com/qwasar-public/only_wood_customer_us_3.csv",sep = '\t',skiprows=1, names=['gender','name','email','age','city','country'])
df3['gender'].unique()
df3['gender'] = df3['gender'].str.replace('string_','')
df3['gender'] = df3['gender'].str.replace('boolean_','')
df3['gender'] = df3['gender'].str.replace('character_','')
genderss = {'0':'Male', '1':'Female','M':'Male'} #Unique qilingan qiymatlarga gender yuklash
df3['gender'] = df3['gender'].replace(genderss) 
df3['gender'].unique()
df3['name'].unique()
df3['name'] = df3['name'].str.replace('string_','')
df3['name'] = df3['name'].str.replace('\W',' ')
df3['name'] = df3['name'].str.title()
name_d = df3['name'].str.split(expand = True)
name_d[2] == None
name_d = df3['name'].str.split(expand=True)
df3['first_name'],df3['last_name'] = name_d[0],name_d[1]
df3.drop('name', axis = 1, inplace = True)
df3['email'] = df3['email'].str.replace('string_','')
df3['email'] = df3['email'].str.lower()
df3.head(5)
df3['age'] = df3['age'].str.replace('\D','')
df3['age']
df3['city'] = df3['city'].str.replace('_',' ')
df3['city'] = df3['city'].str.replace('-',' ')
df3['city'] = df3['city'].str.title()
df3['country'] = 'USA'
df3['city'] = df3['city'].str.replace('String ','')
df3.head(10)
```
